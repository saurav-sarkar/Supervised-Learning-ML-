PS: Let us build a model to predict the CTC offered to a candidate based on Last CTC and Total years of work experience

```python
import pandas as pd
```

```python
# Step 1: Importing the data
ctc = pd.read_csv(r"E:/Dataset/CTCdata.csv")
ctc.head()
```

```python
# Create the subset of data to include only the requisit col
df = ctc[['CTCoffered','LastCTC','Total years of work exp']]
df
```

```python
# Step 2: Identifying X and Y (i.e dependent and independent variable)
Y = df[['CTCoffered']]

X = df[['LastCTC','Total years of work exp']]

# Step 3: To split the data into training and test data
from sklearn.model_selection import train_test_split
 
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, train_size = 0.8, random_state=9)
len(X_train), len(X_test), len(Y_train), len(Y_test)
```

```python
# Step 4: Creating the model using on the training data set
# step a: Create a model object 
from sklearn.linear_model import LinearRegression
lr = LinearRegression()

# step b: Fit the model object into training data to build a model
model = lr.fit(X_train,Y_train)
model
```

```python
#Step 5: Find the values of slope, intercept and R Square (Applicable only when we build Regression based model)
# For linear model R-sq >=75

#Slop(m)
model.coef_
```

```python
#Constant(C)
model.intercept_
```

```python
# To calculate the R-sq value for model
rsq = model.score(X_train, Y_train)
rsq
```

```python
# Step 6: Predict the values on test data using your model
Y_test['Predicted CTC'] = model.predict(X_test)
Y_test
```

```python
# Step 7: Calculate the RMSE value from test data. RMSE - Root <- Mean <- Square <- Error
Y_test['Sqerror'] = (Y_test['CTCoffered'] - Y_test['Predicted CTC']) **2
Y_test.head()
```

```python
#Error Mean
Error_mean = Y_test['Sqerror'].mean()
Error_mean
```
